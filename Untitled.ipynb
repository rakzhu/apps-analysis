```python
# Handling missing values in Google Play Store data

# Fill missing 'Rating' values with median of the column
google_df['Rating'] = google_df['Rating'].fillna(google_df['Rating'].median())

# Fill missing 'Type', 'Content Rating', 'Current Ver', and 'Android Ver' with 'Unknown'
google_df['Type'] = google_df['Type'].fillna('Unknown')
google_df['Content Rating'] = google_df['Content Rating'].fillna('Unknown')
google_df['Current Ver'] = google_df['Current Ver'].fillna('Unknown')
google_df['Android Ver'] = google_df['Android Ver'].fillna('Unknown')

# Removing duplicates
google_df = google_df.drop_duplicates()

# Verify the changes
google_df.isnull().sum(), google_df.duplicated().sum()
```

```python
print(google_df['Price'].unique())
```

```python
google_df = google_df[google_df['Price'].str.match(r'^\$?\d+(\.\d{1,2})?$', na=False)]
```

```python
google_df['Price'] = google_df['Price'].replace({r'\$': '', r',': ''}, regex=True).astype(float)
```

```python
# Remove currency symbol ($) and commas, then convert to float
google_df['Price'] = google_df['Price'].replace({r'\$': '', r',': ''}, regex=True).astype(float)

# Check for any missing values after conversion
google_df['Price'].isnull().sum()
```

```python
# Remove non-English apps by checking if the name contains English characters
apple_df = apple_df[apple_df['track_name'].apply(lambda x: x.isascii())]
google_df = google_df[google_df['App'].apply(lambda x: x.isascii())]
```

```python
# Drop rows where critical columns have missing values
apple_df = apple_df.dropna(subset=['track_name', 'price', 'rating_count_tot'])
google_df = google_df.dropna(subset=['App', 'Category', 'Price'])
```

```python
#visualize the rating distribution for both stores to check for trends.

import matplotlib.pyplot as plt
import seaborn as sns

# Set figure size
plt.figure(figsize=(12, 5))

# Plot histogram for Google Play Store
plt.subplot(1, 2, 1)
sns.histplot(google_df['Rating'].dropna(), bins=20, kde=True, color='blue')
plt.title('Google Play Store Rating Distribution')
plt.xlabel('Rating')
plt.ylabel('Count')

# Plot histogram for Apple App Store
plt.subplot(1, 2, 2)
sns.histplot(apple_df['user_rating'].dropna(), bins=20, kde=True, color='red')
plt.title('Apple App Store Rating Distribution')
plt.xlabel('Rating')
plt.ylabel('Count')

# Show plots
plt.tight_layout()
plt.show()
```

```python
#which app categories have the best ratings in each store
# Calculate average rating per category
google_category_ratings = google_df.groupby('Category')['Rating'].mean().sort_values(ascending=False)
apple_category_ratings = apple_df.groupby('prime_genre')['user_rating'].mean().sort_values(ascending=False)

# Plot the results
plt.figure(figsize=(14, 6))

# Google Play Ratings
plt.subplot(1, 2, 1)
google_category_ratings[:10].plot(kind='bar', color='blue', alpha=0.7)
plt.title('Top 10 Categories by Avg Rating (Google Play)')
plt.xlabel('Category')
plt.ylabel('Avg Rating')
plt.xticks(rotation=45)

# Apple App Store Ratings
plt.subplot(1, 2, 2)
apple_category_ratings[:10].plot(kind='bar', color='red', alpha=0.7)
plt.title('Top 10 Categories by Avg Rating (Apple App Store)')
plt.xlabel('Category')
plt.ylabel('Avg Rating')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
#Most Popular App Categories (Based on Downloads/Installs) which categories have the most downloads, as these indicate market demand
# Convert 'Installs' to numeric in Google Play dataset
google_df['Installs'] = google_df['Installs'].str.replace('[+,]', '', regex=True).astype(int)

# Sum total installs per category
google_category_installs = google_df.groupby('Category')['Installs'].sum().sort_values(ascending=False)

# Count total ratings per category (popularity indicator) in Apple dataset
apple_category_popularity = apple_df.groupby('prime_genre')['rating_count_tot'].sum().sort_values(ascending=False)

# Plot the results
plt.figure(figsize=(14, 6))

# Google Play Popularity
plt.subplot(1, 2, 1)
google_category_installs[:10].plot(kind='bar', color='blue', alpha=0.7)
plt.title('Top 10 Categories by Installs (Google Play)')
plt.xlabel('Category')
plt.ylabel('Total Installs')
plt.xticks(rotation=45)

# Apple App Store Popularity
plt.subplot(1, 2, 2)
apple_category_popularity[:10].plot(kind='bar', color='red', alpha=0.7)
plt.title('Top 10 Categories by Total Ratings (Apple App Store)')
plt.xlabel('Category')
plt.ylabel('Total Ratings')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
#Price vs. Rating Analysis analyze whether paid apps tend to have higher ratings than free apps. 
#This helps businesses decide whether to monetize apps upfront or rely on ads/in-app purchases.

plt.figure(figsize=(14, 6))

# Google Play Store: Price vs. Rating
plt.subplot(1, 2, 1)
sns.scatterplot(data=google_df, x='Price', y='Rating', alpha=0.5, color='blue')
plt.xlim(0, 50)  # Limit price range for better visualization
plt.title('Google Play: Price vs Rating')
plt.xlabel('Price ($)')
plt.ylabel('Rating')

# Apple App Store: Price vs. Rating
plt.subplot(1, 2, 2)
sns.scatterplot(data=apple_df, x='price', y='user_rating', alpha=0.5, color='red')
plt.xlim(0, 50)  
plt.title('Apple App Store: Price vs Rating')
plt.xlabel('Price ($)')
plt.ylabel('Rating')

plt.tight_layout()
plt.show()
```

```python
#Most Popular App Categories analyze which app categories have the most installs (Google Play) and highest number of ratings (Apple App Store). 
#This helps businesses identify high-demand categories.

plt.figure(figsize=(14, 6))

# Google Play: Most Installed Categories
google_category_installs = google_df.groupby('Category')['Installs'].sum().sort_values(ascending=False).head(10)
plt.subplot(1, 2, 1)
sns.barplot(x=google_category_installs.values, y=google_category_installs.index, palette="Blues_r")
plt.title("Top 10 Google Play Categories by Installs")
plt.xlabel("Total Installs")
plt.ylabel("Category")

# Apple Store: Most Rated Categories
apple_category_ratings = apple_df.groupby('prime_genre')['rating_count_tot'].sum().sort_values(ascending=False).head(10)
plt.subplot(1, 2, 2)
sns.barplot(x=apple_category_ratings.values, y=apple_category_ratings.index, palette="Reds_r")
plt.title("Top 10 Apple App Store Categories by Ratings")
plt.xlabel("Total Ratings")
plt.ylabel("Category")

plt.tight_layout()
plt.show()
```

```python
#Price Distribution Analysis
#analyze the price distribution of paid apps on both stores. This helps us understand pricing trends and what users are willing to pay

plt.figure(figsize=(14, 6))

# Google Play: Price Distribution
plt.subplot(1, 2, 1)
sns.histplot(google_df[google_df['Price'] > 0]['Price'], bins=30, kde=True, color='blue')
plt.title("Price Distribution of Paid Apps (Google Play)")
plt.xlabel("Price (USD)")
plt.ylabel("Number of Apps")

# Apple Store: Price Distribution
plt.subplot(1, 2, 2)
sns.histplot(apple_df[apple_df['price'] > 0]['price'], bins=30, kde=True, color='red')
plt.title("Price Distribution of Paid Apps (Apple App Store)")
plt.xlabel("Price (USD)")
plt.ylabel("Number of Apps")

plt.tight_layout()
plt.show()
```

```python
#Correlation Between App Ratings and Price, if higher-priced apps receive better ratings or if price has no effect on ratings.
plt.figure(figsize=(14, 6))

# Google Play: Price vs Rating
plt.subplot(1, 2, 1)
sns.scatterplot(x=google_df[google_df['Price'] > 0]['Price'], y=google_df[google_df['Price'] > 0]['Rating'], color='blue')
plt.title("Price vs Rating (Google Play)")
plt.xlabel("Price (USD)")
plt.ylabel("Rating")
plt.grid(True)

# Apple Store: Price vs Rating
plt.subplot(1, 2, 2)
sns.scatterplot(x=apple_df[apple_df['price'] > 0]['price'], y=apple_df[apple_df['price'] > 0]['user_rating'], color='red')
plt.title("Price vs Rating (Apple App Store)")
plt.xlabel("Price (USD)")
plt.ylabel("Rating")
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
#Analyzing App Size vs. Rating, if app size affects user ratings. Do larger apps get higher ratings, or do users prefer smaller apps?

plt.figure(figsize=(14, 6))

# Google Play: Size vs Rating
plt.subplot(1, 2, 1)
sns.scatterplot(x=google_df['Size'], y=google_df['Rating'], color='blue', alpha=0.5)
plt.title("App Size vs Rating (Google Play)")
plt.xlabel("Size (MB)")
plt.ylabel("Rating")
plt.grid(True)

# Apple Store: Size vs Rating
plt.subplot(1, 2, 2)
sns.scatterplot(x=apple_df['size_bytes'] / 1e6, y=apple_df['user_rating'], color='red', alpha=0.5)
plt.title("App Size vs Rating (Apple App Store)")
plt.xlabel("Size (MB)")
plt.ylabel("Rating")
plt.grid(True)

plt.tight_layout()
plt.show()
```
